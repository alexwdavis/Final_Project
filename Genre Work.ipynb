```python
genre=[]
for i in np.array(df['Genre'].apply(lambda x: x.split(','))):
    for j in i:
        genre.append(j.strip())
```

```python
set(genre)
```

```python
# define columns first
for i in set(genre):
    df[i]= pd.Series()

# populating colmuns 
for x in df['Genre']:
    print(x.split(','))
```

```python
df['Crime'][0] = 1
```

```python
# populating colmuns 
for i,x in enumerate(df['Genre']):
    for words in x.split(','):
        if words.strip() in set(genre):
            df[words.strip()][i] = 1
```
