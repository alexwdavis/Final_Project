```python
df[['Genre', 'Director', 'Box Office','Language','Country','IMDB', 'Metacritic','Rank','Awards', 'Worldwide Box Office', 'Domestic Box Office', 'International Box Office']]
```

```python
df['Language']=df['Language'].fillna("")
```

```python
language=[]
for i in np.array(df['Language'].apply(lambda x: x.split(','))):
    for j in i:
        language.append(j.strip())
```

```python
set(language)
```

```python
# define columns first
for i in set(language):
    df[i]= pd.Series()

# populating colmuns 
for x in df['Language']:
    print(x.split(','))
```

```python
df['English'][0] = 1
```

```python
# populating colmuns 
for i,x in enumerate(df['Language']):
    for words in x.split(','):
        if words.strip() in set(language):
            df[words.strip()][i] = 1
```
